```python
import requests
import pandas as pd
```

```python
APP_ID = '1098649060595452317'
REQUEST_URL = 'https://app.rakuten.co.jp/services/api/IchibaItem/Search/20170706'
```

```python
params = {
    'applicationId': APP_ID,
    'format': 'json',
    'keyword': 'フルーティー アイドル'

}
```

```python
res = requests.get(REQUEST_URL, params)
result = res.json()
```

```python
items = result['Items']
len(items)
```

```python
items = [item['Item'] for item in items]
items[0]
```

```python
df = pd.DataFrame(items)
```

```python
df.columns
```

```python
columns = ['itemCaption', 'itemCode', 'itemName', 'availability', 'catchcopy', 'itemPrice', 'itemUrl', 'reviewAverage', 'reviewCount','shopCode', 'shopName', 'shopUrl']
```

```python
df = df[columns]
```
